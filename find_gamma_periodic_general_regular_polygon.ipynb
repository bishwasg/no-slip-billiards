```python
from pylab import *
import math
import numpy as np

epsilon= 0.0001 #tolerance
n = 8 #periodicity
s = 4 #no. of sides in a regular polygon or 2 for the 2D plane strip

delta = 0.001
gamma_range = np.arange(delta,1+delta,delta)
def gamma_periodic(n,s):
     
    #min_error = 100
    periodic_gamma = []
    min_error = []
    error = np.array([None]*len(gamma_range))
    for i in range(len(gamma_range)):
        print(i)
        #find angles to rotate the velocity vector after each collision 
        if s == 2:
            theta_range = np.array([(k+1)*2*math.pi/s for k in range(s-1)])  
        else:
            theta_range = np.array([(k+1)*math.pi/s for k in range(s-1)])  
        
        #initialize different states possible at each collision 
        cases = (s-1)**n
        identity = np.array([[1,0,0],
                           [0,1,0],
                           [0,0,1]])
        states = [identity for j in range(cases)]
        
        #run the simulation for this particular gamma
        gamma = gamma_range[i]
        col_count = 0
        while col_count < n:
            states = next_collision(states,theta_range,gamma)
            col_count += 1   
        #find error for each state for this gamma
        error[i] = min([find_error(temp) for temp in states])
    
    temp_index  = sorted(np.argpartition(error, 5)[:5])
    print(error[temp_index])
    print(gamma_range[temp_index])
    min_error = np.append(min_error, error[temp_index])
    periodic_gamma = np.append(periodic_gamma, gamma_range[temp_index])
    
    #min_error = np.append(min_error, error[0])
    #periodic_gamma = np.append(periodic_gamma, gamma_range[0])
    
    #for index in range(1,len(gamma_range)-1,1):
        #if error[index-1] > error[index] and error[index+1] > error[index]:
            #min_error = np.append(min_error, error[index])
            #periodic_gamma = np.append(periodic_gamma, gamma_range[index])
    #get the first and last error
    #min_error = np.append(min_error, error[-1])
    #periodic_gamma = np.append(periodic_gamma, gamma_range[-1])
                                      
    return [min_error, periodic_gamma]
            
        

def find_error(x):
    error = 0 
    for i in range(3):
        for j in range(3):
            if i == j: 
                error += (x[i][j]-1)**2 
            else:
                error += x[i][j]**2
    return error

def rotation(theta):
    return np.array([[1,0,0],
                        [0,cos(theta), -sin(theta)],
                        [0,sin(theta), cos(theta)]])
def reflection(gamma):
    return np.array([[-(1-gamma**2)/(1+gamma**2), -2*gamma/(1+gamma**2),0],
                                    [-2*gamma/(1+gamma**2),(1-gamma**2)/(1+gamma**2),0],
                                   [0,0,-1]])

def new_state(current_state, theta, gamma):
    T_gamma = reflection(gamma)
    R_theta = rotation(theta)
    return np.matmul(np.matmul(R_theta, T_gamma), current_state) 

def next_collision(states, theta_range, gamma):
    n_states = len(states)
    n_theta = len(theta_range)  
    for i in range(n_theta):
        if n_states == n_theta:
            states[i] = new_state(states[i],theta_range[i],gamma)         
        else:  
            bin_len = n_states/n_theta
            start_index = bin_len*i
            end_index = bin_len*(i+1) 
            for j in range(start_index,end_index):
                states[j] = new_state(states[j],theta_range[i],gamma)
            states[start_index:end_index] = next_collision(states[start_index:end_index],theta_range,gamma)
    return states

gamma_periodic(n,s)


#gamma for no-slip strip with period 24
#error: array([1.00464482e-05, 9.70721528e-06, 6.17517661e-07, 6.41013608e-06,1.32963094e-06])
#gamma values: array([0.1317, 0.2679, 0.4142, 0.5774, 0.7673])

#equilateral triangle
#found 1/math.sqrt(2) = 0.707, but also found other values 
#error: [array([8.00000000e+00, 1.45936158e-09, 3.54147045e-09, 2.35515214e-11, 2.41451843e-10, 1.98925888e-09, 7.05060089e+00])
#gamma values:  array([0.     , 0.20467, 0.27345, 0.31739, 0.37035, 0.70711, 0.99999])]
#However, any gamma other than 1/sqrt(2) didn't give periodic orbits, other cases might not be physically permissible

```
